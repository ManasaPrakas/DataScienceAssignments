```python
# Calculate percentages for categorical features

sex_counts = data['sex '].value_counts()
sex_percentages = (sex_counts / len(data)) * 100
print(sex_percentages)

chest_pain_counts = data['chest pain type'].value_counts()
chest_pain_percentages = (chest_pain_counts / len(data)) * 100
print(chest_pain_percentages)

fasting_blood_sugar_counts = data['fasting blood sugar'].value_counts()
fasting_blood_sugar_percentages = (fasting_blood_sugar_counts / len(data)) * 100
print(fasting_blood_sugar_percentages)

resting_electrocardiographic_counts = data['resting electrocardiographic results'].value_counts()
resting_electrocardiographic_percentages = (resting_electrocardiographic_counts / len(data)) * 100
print(resting_electrocardiographic_percentages)

excercise_induced_angina_counts = data['exercise induced angina'].value_counts()
excercise_induced_angina_percentages = (excercise_induced_angina_counts / len(data)) * 100
print(excercise_induced_angina_percentages)

ST_segment_counts = data['ST segment'].value_counts()
ST_segment_percentages = (ST_segment_counts / len(data)) * 100
print(ST_segment_percentages)

thal_counts = data['thal'].value_counts()
thal_percentages = (thal_counts / len(data)) * 100
print(thal_percentages)

heart_disease_counts = data['heart disease'].value_counts()
heart_disease_percentages = (heart_disease_counts / len(data)) * 100
print(heart_disease_percentages)
```

Summary of Initial Analysis

Missing Values:

There are no missing values in the dataset.

Descriptive Statistics:

The dataset contains 270 rows.

The average age is around 54.43 years.

About 67.78% of the participants are male.

The chest pain type is predominantly of type 3 (non-anginal pain) and type 4 (asymptomatic).

The average resting blood pressure is around 131.34 mm Hg.

The average serum cholesterol level is 249.66 mg/dl.

Only about 14.81% of the participants have a fasting blood sugar level > 120 mg/dl.

The resting electrocardiographic results are distributed among categories 0 and 2.

The average maximum heart rate achieved is around 149.68 bpm.

About 32.96% of the participants experience exercise-induced angina.

The average 'oldpeak' value (ST depression induced by exercise) is 1.05.

The average number of major vessels colored by fluoroscopy is around 0.67.

The 'thal' variable has three unique values with an average around 4.7.

The target variable 'heart disease' has a mean value of 1.44, indicating that the dataset has more patients with heart disease (1) compared to those without (2).

**Heart Disease affect based on Gender, Age**

```python
gender_heart_disease = data.groupby('sex ')['heart disease'].value_counts(normalize=True).unstack()
gender_heart_disease.plot(kind='bar', stacked=True)
plt.title('Heart Disease by Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Proportion')
plt.show()
```

Based on above graph, Female is more prone to get heart disease than Male.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='heart disease', y='age', data=data)
plt.title('Age vs Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 2 = No)')
plt.ylabel('Age')
plt.show()
```

Based on the above plot Heart disease does not have any effect on age.

**Visualize Distribution**

Analyzing Numerical Features

```python
plt.figure(figsize=(10, 6))
numerical_features = ['age', 'resting blood pressure', 'serum cholestoral', 'max heart rate', 'oldpeak']

# Plot distribution of numerical features
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(3, 2, i)
    sns.histplot(data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Analyze Categorical Variables

```python
plt.figure(figsize=(20, 15))

# List of categorical features
categorical_features = ['sex ', 'chest pain type', 'fasting blood sugar',
                        'resting electrocardiographic results', 'exercise induced angina',
                        'ST segment', 'major vessels', 'thal', 'heart disease']

# Plot distribution of categorical features
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 3, i)
    sns.countplot(x=data[feature])
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

**Correlation Analysis**

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

**Target Variable Analysis**

```python
# Distribution of heart disease status
plt.figure(figsize=(8, 6))
sns.countplot(x='heart disease', data=data)
plt.title('Distribution of Heart Disease Status')
plt.xlabel('Heart Disease (1 = Yes, 2 = No)')
plt.ylabel('Count')
plt.show()

# Relationship between age and heart disease
plt.figure(figsize=(10, 6))
sns.boxplot(x='heart disease', y='age', data=data)
plt.title('Age vs Heart Disease')
plt.xlabel('Heart Disease (1 = Yes, 2 = No)')
plt.ylabel('Age')
plt.show()

# Relationship between other features and heart disease
for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='heart disease', y=feature, data=data)
    plt.title(f'{feature} vs Heart Disease')
    plt.xlabel('Heart Disease (1 = Yes, 2 = No)')
    plt.ylabel(feature)
    plt.show()
```
